Seaborn

Seaborn is a Python data visualization library based on Matplotlib, providing a high-level interface for drawing attractive and informative statistical graphics. It simplifies the creation of complex plots such as heatmaps, violin plots, and pair plots with just a few lines of code.

Line Plot

A line plot is a basic plot that displays data points connected by straight line segments.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data
x = [1, 3, 5, 7, 9]
y = [2, 4, 6, 8, 10]

# Plotting a line plot
sns.lineplot(x=x, y=y)
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Styling Line Plot

You can style line plots by customizing parameters such as line color, markers, and linestyle.


```python

sns.lineplot(x=x, y=y, color='green', marker='o', linestyle='--')
plt.title('Styled Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

Histogram

A histogram in Python can be created using libraries like Matplotlib or Seaborn to visualize the frequency distribution of a dataset. The plt.hist() function in Matplotlib or sns.histplot() in Seaborn are commonly used for this purpose.








```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset from Seaborn
tips = sns.load_dataset("tips")

# Plotting a histogram with KDE
sns.histplot(data=tips, x="total_bill", bins=17, kde=True)
plt.title('Histogram of Total Bill Amounts with KDE')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

Bar Plot

A bar plot displays categorical data with rectangular bars where the height represents the value.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  # Import numpy for using np.mean

# Load the tips dataset from Seaborn
tips = sns.load_dataset("tips")

# Plotting a bar plot
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", estimator=np.mean)
plt.title('Average Total Bill by Day and Gender')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill ($)')
plt.show()
```

Heatmap

A heatmap visualizes data using a color gradient to represent the numerical values in a matrix format.

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset from Seaborn
flights = sns.load_dataset("flights")

# Pivot the data to create a pivot table for heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Plotting a heatmap
sns.heatmap(data=flights_pivot, cmap='coolwarm', linecolor='white', linewidth=1)
plt.title('Heatmap of Flight Passengers')
plt.show()
```

Scatter Plot

A scatter plot shows the relationship between two numerical variables by placing points on a Cartesian plane.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'tips' dataset is defined and loaded correctly
# Example dataset loading:
tips = sns.load_dataset("tips")

# Plotting a scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time")
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

KDE Plot

A KDE plot (Kernel Density Estimate) displays the distribution of a single variable or the relationship between two variables using contours.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'tips' dataset is defined and loaded correctly
# Example dataset loading:
tips = sns.load_dataset("tips")

# Plotting a KDE plot
sns.kdeplot(data=tips, x="total_bill", hue="sex", fill=True)
plt.title('KDE Plot of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Density')
plt.show()
```

Rug Plot

A rug plot displays data points as ticks on an axis to show the distribution of data points.

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'tips' dataset is defined and loaded correctly
# Example dataset loading:
tips = sns.load_dataset("tips")

# Plotting a rug plot
sns.rugplot(data=tips, x="total_bill", height=0.3)
plt.title('Rug Plot of Total Bill')
plt.xlabel('Total Bill ($)')
plt.show()
```

Box Plot

A box plot summarizes the distribution of a numerical variable across different categories, showing quartiles and potential outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'tips' dataset is defined and loaded correctly
# Example dataset loading:
tips = sns.load_dataset("tips")

# Plotting a box plot
sns.boxplot(data=tips, x="day", y="total_bill", hue="time")
plt.title('Box Plot of Total Bill by Day and Time')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

Violin Plot

A violin plot combines aspects of a box plot and a KDE plot to show the distribution of data across different categories

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'tips' dataset is defined and loaded correctly
# Example dataset loading:
tips = sns.load_dataset("tips")

# Plotting a violin plot
sns.violinplot(data=tips, x="day", y="total_bill", hue="time", split=True)
plt.title('Violin Plot of Total Bill by Day and Time')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.show()
```

Pair Plot

A pair plot creates a grid of scatter plots and histograms for pairs of variables in a dataset to show relationships and distributions across the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'tips' dataset is defined and loaded correctly
# Example dataset loading:
tips = sns.load_dataset("tips")

# Plotting a pair plot
sns.pairplot(data=tips, vars=["total_bill", "tip", "size"], hue="sex")
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```
